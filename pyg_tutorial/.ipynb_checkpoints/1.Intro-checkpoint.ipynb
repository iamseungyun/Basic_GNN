# NetworkX Tutorial

## Setup

```python
import networkx as nx
```

## Graph

```python
# Create an undirected graph
G = nx.Graph()
print(G.is_directed())

# Create a directed graph
H = nx.DiGraph()
print(H.is_directed())

# Add graph level attribute
G.graph["Name"] = "Bar"
print(G.graph)
```

## Node

```python
# Add one node with node level attributes
G.add_node(0, feature=5, label=0)

# Get attributes of the node 0
node_0_attr = G.nodes[0]
print("Node 0 has the attributes {}".format(node_0_attr))
```

```python
# Add multiple nodes with attributes
G.add_nodes_from([
    (1, {"feature": 1, "label": 1}),
    (2, {"feature": 2, "label": 2})
])

# Loop through all the nodes
# Set data=True will return node attributes
for node in G.nodes(data=True):
    print(node)

# Get number of nodes
num_nodes = G.number_of_nodes()
print("G has {} nodes".format(num_nodes))
```

## Edge

```python
# Add one edge with edge weight 0.5
G.add_edge(0,1, weight=0.5)

# Get attributes of the edge (0,1)
edge_0_1_attr = G.edges[(0,1)]
print("Edge (0,1) has the attributes {}".format(edge_0_1_attr))
```

```python
# Add multiple edges with edge weights
G.add_edges_from([
    (1, 2, {"weight": 0.3}),
    (2, 0, {"weight": 0.1})
])

# Loop through all the edges
# Here there is no data=True, so only the edge will be returned
for edge in G.edges():
    print(edge)

# Get number of edges
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges))
```

## Visualization

```python
# Draw the graph
nx.draw(G, with_labels=True)
```

## Node Degree and Neighbor

```python
node_id = 1

# Degree of node 1
print("Node {} has degree {}".format(node_id, G.degree[node_id]))

# Get neighbor of node 1
for neighbor in G.neighbors(node_id):
    print("Node {} has neighbor {}".format(node_id, neighbor))
```

## Other Functionalities

```python
num_nodes = 4

# Create a new path like graph and change it to a directed graph
G = nx.DiGraph(nx.path_graph(num_nodes))
nx.draw(G, with_labels=True)

# Get the PageRank
pr = nx.pagerank(G, alpha=0.8)
pr
```

## Visualization

```python
# Helper function for visualization
%matplotlib inline
import torch
import networkx as nx
import matplotlib.pyplot as plt

# Visualize function
def visualize(h, color, epoch=None, loss=None, accuracy=None):
    plt.figure(figsize=(7,7))
    plt.xticks([])
    plt.yticks([])

    if torch.is_tensor(h):
        h = h.detach().cpu().numpy()
        plt.scatter(h[:,0], h[:,1], s=140, c=color, cmap="Set2")
        if epoch is not None and loss is not None and accuracy['train'] is not None and accuracy['val'] is not None:
            plt.xlabel((f'Epoch: {epoch}, Loss: {loss.item():.4f} \n'
                        f'Training Accuracy: {accuracy["train"]*100:.2f}% \n'
                        f'Validation Accuracy: {accuracy["val"]*100:.2f}%'),
                        fontsize=16)
    else:
        nx.draw_networkx(h, pos=nx.spring_layout(h, seed=42), with_labels=False, node_color=color, cmap="Set2")
    plt.show()
```

## Implement GNN

```python
import torch
from torch.nn import Linear
from torch_geometric.nn import GCNConv

class GCN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes, num_layers):
        super().__init__()
        torch.manual_seed(1234)

        self.convs = torch.nn.ModuleList()

        self.convs.append(GCNConv(input_dim, hidden_dim))
        for _ in range(num_layers-2):
            self.convs.append(GCNConv(hidden_dim, hidden_dim))
        self.convs.append(GCNConv(hidden_dim,2))

        self.classifier = Linear(2, num_classes)

    def forward(self, x, edge_index):
        for conv in self.convs:
            x = conv(x, edge_index)
            x = torch.tanh(x)

        embeddings = x
        out = self.classifier(embeddings)
        return out, embeddings

model = GCN(input_dim=dataset.num_features, hidden_dim=4, num_classes=dataset.num_classes, num_layers=3)
print(model)
```

```python
model = GCN(num_classes=dataset.num_classes, input_dim=dataset.num_features, hidden_dim=4, num_layers=3)

_, h = model(data.x, data.edge_index)
print(f'Embedding shape: {list(h.shape)}')

visualize(h, color=data.y)
```

```python
import time

model = GCN(num_classes=dataset.num_classes, input_dim=dataset.num_features, hidden_dim=4, num_layers=3)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

def train(data):
    for epoch in range(500):
        optimizer.zero_grad()
        out, h = model(data.x, data.edge_index)
        loss = criterion(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()

        accuracy={}
        train_predicted_classes = torch.argmax(out[data.train_mask], axis=1)
        train_target_classes = data.y[data.train_mask]
        accuracy['train'] = torch.mean(
            torch.where(train_predicted_classes == train_target_classes, 1, 0).float()
        )

        predicted_classes = torch.argmax(out, axis=1)
        target_classes = data.y
        accuracy['val'] = torch.mean(
            torch.where(predicted_classes == target_classes, 1, 0).float()
        )

        if epoch % 25 == 0:
            visualize(h, color=data.y, epoch=epoch, loss=loss, accuracy=accuracy)
            time.sleep(0.5)

train(data)
```
